```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

# Wykres Liniowy

```python
plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.xlabel("axis X")
plt.ylabel("axis Y")
plt.title("Tytu≈Ç wykresu")
_ = plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel("axis X")
plt.ylabel("axis Y")
plt.title("Tytu≈Ç wykresu")
_ = plt.legend()
```

```python
x = np.arange(-3.2, 3.2, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

# Twoerzenie wielu wykres√≥w

```python
plt.subplot(121)#1 - liczba wielrszy wykres√≥w, 2 - liczba kolumn wykres√≥w, 1 - nr wykresu
plt.plot([1, 2, 3], [3, 2, 5])

plt.subplot(122)#1 - liczba wielrszy wykres√≥w, 2 - liczba kolumn wykres√≥w, 2 - nr wykresu
plt.plot([1, 2, 3], [2, 4, 3], 'r--')
```

```python
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)#1 - liczba wielrszy wykres√≥w, 3 - liczba kolumn wykres√≥w, 1 - nr wykresu
plt.plot([1, 2, 3], [3, 2, 5])

plt.subplot(132)#1 - liczba wielrszy wykres√≥w, 3 - liczba kolumn wykres√≥w, 2 - nr wykresu
plt.plot([1, 2, 3], [2, 4, 3], 'r--')

plt.subplot(133)#1 - liczba wielrszy wykres√≥w, 3 - liczba kolumn wykres√≥w, 3 - nr wykresu
plt.plot([1, 2, 3], [2, 4, 3], 'bo-')
```

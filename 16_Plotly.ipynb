#Plotly

```python
import plotly
import plotly.graph_objects as go
plotly.__version__
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4])
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title={'text':"Wykres słupkowy"})
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title_text="Wykres słupkowy1")
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title_text="Wykres słupkowy1")
)
fig.write_html("bar.html")#zapisanie interaktywnego wykresu do pliku html
fig.show()
```

##Wykres kołowy i pierścieniow

```python
from seaborn import load_dataset
df = load_dataset("diamonds")
df.head()
```

```python
df.info()
```

```python
dff = df.cut.value_counts()
dff
```

```python
dff = df.cut.value_counts()
dff = dff.reset_index()
dff.columns=['type', 'count']
dff
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff["type"], values=dff["count"])
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff["type"], values=dff["count"]),
    layout=go.Layout(title_text="Rozklad zmiennej Cut", width=800)
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff["type"], values=dff["count"], hole=0.7),
    layout=go.Layout(title_text="Rozklad zmiennej Cut", width=800)
)
fig.show()
```

##Diagram Sankey'a

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2024', 'Churn_2024', 'Nonchurn_2025', 'Churn_2025']),
                link=dict(source=[0, 0, 1, 1],#index odpowiadajacy etykiecie  ^ SOURCE
                          target=[2, 3, 2, 3],#Target
                          value=[65, 12, 18, 5]))]
fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2023', 'Churn_2023', 'Nonchurn_2024', 'Churn_2024', 'Nonchurn_2025', 'Churn_2025']),
                link=dict(source=[0, 1,  2, 2, 3, 3],#index odpowiadajacy etykiecie  ^ SOURCE
                          target=[2, 3, 4, 5, 4, 5],#Target
                          value=[77, 23, 65, 12, 18, 5]))]
fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

## Wykres świecowy

```python
from os import name
def fetch_financial_data(company="AMZN"):
  """Fetch stock data from stooq.pl service"""
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df = df[df.Date>"2024-01-01"]
df.head()
```

```python
fig = go.Figure(
    data=go.Candlestick(x=df.Date, open=df.Open, high=df.High, low=df.Low, close=df.Close)
)
fig.show()
```
